# Matrix Operations and Functions
Recall that arithmetic array operations +, -, /, * and ** are performed elementwise on NumPy arrays[link text](https://)

```
import numpy as np
import scipy.linalg as la
from numpy.linalg import matrix_power as mpow


#------- Arithmetic Operations
M = np.array([[3,4],[-1,5]])
print(M)

print(M * M) #The M_Matrix Array the multiply.

#------- Matrix Multiplication
print(M @ M) 
#We use the @ operator to do matrix multiplication with NumPy arrays:

A = np.array([[1,3],[-1,7]])
print(A)

B = np.array([[5,2],[1,2]])
print(B)

I = np.eye(2)
print(I)

#The finde Identity Matrix of A,B.
print(2*I + 3*A - A@B)

#------- Matrix Powers
M = np.array([[3,4],[-1,5]])
print(M)

print(mpow(M,2))

print(mpow(M,5))

print(M @ M @ M @ M @ M)
#Compare with the matrix multiplcation operator

#------- Tranpose
print(M.T)

print(M @ M.T)

#------- Inverse
A = np.array([[1,2],[3,4]])
print(A)

print(la.inv(A))

#------- Trace
print(np.trace(A))

#------- Determinant
A = np.array([[1,2],[3,4]])
print(A)

print(la.det(A))

```

# Linear Systems: Eigenvalues and Eigenvectors



```
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

A = np.array([[1,0],[0,-2]])
print(A)

results = la.eig(A)
print(results[0])
print(results[1])

eigvals, eigvecs = la.eig(A)
print(eigvals)
print(eigvecs)

eigvals = eigvals.real
print(eigvals)

lambda1 = eigvals[1]
print(lambda1)

v1 = eigvecs[:,1].reshape(2,1)
print(v1)

print(A @ v1)

print(lambda1 * v1)

```

# Linear System: Polynomial Interpolation


```
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

x = np.array([0,3,8])
X = np.column_stack([[1,1,1],x,x**2])
print(X)

y = np.array([6,1,2]).reshape(3,1)
print(y)

a = la.solve(X,y)
print(a)

#--------- One -----------#
xs = np.linspace(0,8,20)
ys = a[0] + a[1]*xs + a[2]*xs**2
plt.plot(xs,ys,x,y,'b.',ms=20)
plt.show()

#--------- Two -----------#

N = 10
x = np.arange(0,N)
y = np.random.randint(0,10,N)
plt.plot(x,y,'r.')
plt.show()

#--------- Three -----------#
X = np.column_stack([x**k for k in range(0,N)])
print(X[:5,:5])

X = np.vander(x,increasing=True)
print(X[:5,:5])

a = la.solve(X,y)

xs = np.linspace(0,N-1,200)
ys = sum([a[k]*xs**k for k in range(0,N)])
plt.plot(x,y,'r.',xs,ys)
plt.show()
```

# Linear System: Fake Noisy Quadratic Data

```
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

a0 = 3
a1 = 5
a2 = 8
N = 1000
x = 2*np.random.rand(N) - 1 # Random numbers in the interval (-1,1)
noise = np.random.randn(N)
y = a0 + a1*x + a2*x**2 + noise
plt.scatter(x,y,alpha=0.5,lw=0);
plt.show()
```
